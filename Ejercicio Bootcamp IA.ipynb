3.- ¿Qué es la interpolación de cadenas? 

La interpolación de cadenas en Python es una técnica para insertar valores de variables en una cadena para crear una nueva cadena. Es una forma más fácil y legible de crear una cadena que contiene valores dinámicos.

Phyton ofrece  varias formas de realizar la interpolación de cadenas:

1.- Usando la sintaxis de llaves {}: Esta es la forma más común de interpolación de cadenas en Python. Se agrega una f antes de la cadena y se encierra el nombre de la variable entre llaves {} dentro de la cadena. Python reemplaza automáticamente las llaves con el valor actual de la variable correspondiente.
Ejemlo: mensaje = f"Hola, mi nombre es {nombre} y tengo {edad} años."

2.- Usando el método format(): Este método permite especificar los valores de las variables en la cadena utilizando llaves {} y luego llamar al método format() para proporcionar los valores.
Ejemplo: mensaje = "Hola, mi nombre es {} y tengo {} años.".format(nombre, edad)

3.-Usando %: Esta es una forma antigua de interpolación de cadenas en Python que todavía es compatible. Es similar a la forma en que se formatean las cadenas en C. 
Emjemplo: mensaje = "Hola, mi nombre es %s y tengo %d años." % (nombre, edad)




```python
# Definición de las variables
nombre = "María"
edad = 32
lugardetrabajo = "Banco Azteca"

# Interpolación de variables
mensaje = f"Hola, ella es {nombre}. Tiene {edad} años y trabaja en {lugardetrabajo}."

# Imprimir el mensaje
print(mensaje)
```

4.1.- ¿En qué se diferencia una lista de un diccionario?

Tanto las listas como los diccionarios son estructuras de datos en Python que se utilizan para almacenar y organizar datos. La principal diferencia entre una lista y un diccionario es la forma en que se accede y se almacenan los elementos.

Una lista es una colección ordenada de elementos que se almacenan en una secuencia. Los elementos de una lista se identifican por su posición en la secuencia, comenzando desde el índice 0. Los elementos de una lista pueden ser de diferentes tipos de datos, como enteros, cadenas, flotantes, booleanos y otras listas.

Ejemplo, una lista de números enteros:
numeros = [1, 2, 3, 4, 5]
Para acceder a un elemento en una lista, se utiliza el operador de corchetes [] seguido del índice del elemento. 

Ahora el concepto de un diccionario: es una colección desordenada de elementos que se almacenan como pares de clave-valor. Cada elemento en un diccionario se identifica por su clave en lugar de una posición. Las claves en un diccionario deben ser únicas y no pueden ser modificadas después de su creación. Los valores en un diccionario pueden ser de diferentes tipos de datos, como enteros, cadenas, flotantes, booleanos y otras listas y diccionarios.

Ejemplo de un diccionario que contiene información sobre una persona:

persona = {"nombre": "María", "edad": 32, "Lugar de trabajo": "Banco Azteca"}

Para acceder a un valor en un diccionario, se utiliza la clave correspondiente en lugar de un índice. Por ejemplo, para acceder al valor de la clave "edad" en el diccionario persona, se utilizaría la sintaxis persona["edad"].

La diferencia entre una lista y un diccionario es que una lista es una colección ordenada de elementos identificados por su posición, mientras que un diccionario es una colección desordenada de elementos identificados por claves únicas.

```python
#Ejemplos:

#Lista
numeros = [1, 2, 3, 4, 5]
print("ejemplo de lista: ",numeros[0])

#Diccionario
persona = {"nombre": "María", "edad": 32, "Lugar de trabajo": "Banco Azteca"}
print(persona)

persona = {"nombre": "María", "edad": 32, "Lugar de trabajo": "Banco Azteca"}
for clave in persona:
    valor = persona[clave]
    print(clave + ": " + str(valor))
```

4.2.- ¿En qué se diferencia un diccionario de una tuppla?

En Python, tanto las listas como las tuplas son estructuras de datos que se utilizan para almacenar y organizar datos, pero hay algunas diferencias importantes entre ellas.

La principal diferencia entre una lista y una tupla es que una lista es mutable, lo que significa que se puede modificar después de su creación, mientras que una tupla es inmutable, lo que significa que no se puede modificar después de su creación.

Otra diferencia importante es que las listas se definen utilizando corchetes [ ], mientras que las tuplas se definen utilizando paréntesis ( ). 

Las listas tienen más métodos incorporados en Python que las tuplas, lo que las hace más flexibles y fáciles de trabajar en ciertas situaciones. Por ejemplo, las listas tienen métodos como append(), extend(), insert(), remove(), pop() y sort(), que se utilizan para agregar, eliminar y ordenar elementos. Las tuplas, por otro lado, solo tienen dos métodos incorporados: count() y index(), que se utilizan para contar el número de ocurrencias de un elemento y para encontrar el índice de un elemento en la tupla.

Las principales diferencias entre una lista y una tupla son que las listas son mutables, se definen utilizando corchetes y tienen más métodos incorporados en Python, mientras que las tuplas son inmutables, se definen utilizando paréntesis y tienen menos métodos incorporados en Python.





```python
#ESCRIBE UN PROGRAMA QUE DEVUELVA TODAS LAS CLAVES DE UN DICCIONARIO EN FORMA DE TUPLA, POR EJEMPLO: {“name”: “Javier”, “age”: 30, “height”: 1.86} -> (“name”, “age”, “height”)


diccionario = {"name": "Javier", "age": 30, "height": 1.86}

# Obtener todas las claves del diccionario en forma de tupla
tupla_claves = tuple(diccionario.keys())

# Imprimir la tupla de claves
print(tupla_claves)

tupla_ = {"name": "Javier", "age": 30, "height": 1.86}
for clave in tupla_:
    valor = tupla_[clave]
    print(clave + ": " + str(valor))
```

¿EN QUÉ SE DIFERENCIA UNA CLASE Y UN OBJETO?. 

En Python, una clase es un tipo de dato que describe un conjunto de atributos y métodos que comparten todos los objetos que pertenecen a esa clase.

La clase es como una plantilla o un molde que define la estructura y el comportamiento que tendrán los objetos que se creen a partir de ella.

Un objeto es una instancia concreta de una clase. Es decir, es un ejemplar que se crea a partir de la plantilla o molde definidos por la clase. Cada objeto tiene sus propios valores de atributos y puede ejecutar los métodos definidos por su clase.

Su diferencia radica en que la clase es la definición abstracta de un conjunto de objetos, mientras que un objeto es una entidad concreta que se crea a partir de esa definición y que tiene su propia identidad, estado y comportamiento





```python
#  7.1.- CREA UNA CLASE CON UN MÉTODO DE INICIALIZACIÓN CON DOS PARÁMETROS, DOS MÉTODOS Y DOS ATRIBUTOS E INSTANCIA UN OBJETO.

class Colaborador:

    def __init__(self):
        self.nombre=input("Ingrese el nombre del colaborador:")
        self.salario=float(input("Ingrese el salario anual:"))

    def imprimir(self):
        print("Nombre:",self.nombre)
        print("Salario:",self.salario)

    def paga_impuestos(self):
        if self.salario>30000:
            print("Debe pagar impuestos")
        else:
            print("No paga impuestos")


# bloque principal

empleado1=Colaborador()
empleado1.imprimir()
empleado1.paga_impuestos()



```

9.- DESCRIBE QUÉ ES UNA EXCEPCIÓN Y EN PYTHON Y PON UN EJEMPLO.


Una excepción en Python es un evento que ocurre durante la ejecución de un programa que interrumpe el flujo normal de las instrucciones. Cuando ocurre un error o una situación inesperada en el programa, se genera una excepción que indica que algo salió mal. Las excepciones pueden ser generadas automáticamente por el intérprete de Python o por el programador mediante la instrucción "raise".

Por ejemplo, si intentamos dividir un número por cero en Python, se generará una excepción del tipo "ZeroDivisionError". Esto sucede porque no es posible realizar la operación matemática de la división entre cero. Para manejar esta excepción, podemos utilizar la instrucción "try-except" en Python, que nos permitirá detectar y manejar el error de manera adecuada sin interrumpir la ejecución del programa.

```python
# 9.1.- Ejemplo de Excepción

num1 = 10
num2 = 0

try:
    resultado = num1 / num2
except ZeroDivisionError:
    print("Error: No es posible dividir un número por cero.")
```
